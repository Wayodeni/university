```python
import re
import csv
import requests


re_item = re.compile(
    r'<div class="widgets-list__item">(.*?)</div>\s*</div>\s*</div>',
    re.DOTALL # Метасимвол . соответствует любому символу, включая \n
)

re_name = re.compile(
    r'<a[\s\S]* class=\"org-widget-header__title-link\">(.*?)</a>',
    re.DOTALL
)

re_address = re.compile(
    r'<span class=\"org-widget-header__meta org-widget-header__meta--location\">[\s\S](.*)',
)

re_phone = re.compile(
    r'<span class="spec__index-inner">Телефон</span></dt>\s*<dd class="spec__value">(.*?)</dd>',
    re.DOTALL
)

re_hours = re.compile(
    r'<span class="spec__index-inner">Часы работы</span></dt>\s*<dd class="spec__value">(.*?)</dd>',
    re.DOTALL
)

re_site = re.compile(
    r'data-url="([^"]+)"'
)

items = []

for page in range(1, 5 + 1):
    print(f"Страница {page}...")

    html = requests.get(f"https://msk.spravker.ru/avtoservisy-avtotehcentry/?page={page}").text

    blocks = re_item.findall(html)

    for block in blocks:
        name_match = re_name.search(block)
        addr_match = re_address.search(block)
        phone_match = re_phone.search(block)
        hours_match = re_hours.search(block)
        site_match = re_site.search(block)

        name = name_match.group(1).strip() if name_match else ""
        address = addr_match.group(1).strip() if addr_match else ""
        phone = phone_match.group(1).strip() if phone_match else ""
        hours = hours_match.group(1).strip() if hours_match else ""

        # сайт хранится в data-url как base64
        if site_match:
            import base64
            site = base64.b64decode(site_match.group(1)).decode("utf-8")
        else:
            site = ""

        items.append([name, address, phone, hours, site])

print(f"Всего собрано: {len(items)}")
items

```

```python
print("Начало записи csv")
with open("spravker.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter=";")
    writer.writerow(["Название", "Адрес", "Телефон", "Часы работы", "Сайт"])
    writer.writerows(items)


print("Запись csv завершена")
```
