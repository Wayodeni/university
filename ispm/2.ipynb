## 4. ДЕТАЛЬНАЯ ХАРАКТЕРИСТИКА СИСТЕМ-АНАЛОГОВ

### 4.1. Polar View

#### 4.1.1. Общее описание

Polar View - специализированный веб-сервис оперативной информации о морском льде, ориентированный на Арктику и Антарктику. Он управляется British Antarctic Survey и предназначен прежде всего для практических задач навигации и ледовой обстановки, а не как универсальная платформа для пользовательского машинного обучения. В сервисе доступны SAR-изображения, ледовые карты, данные по концентрации льда и отслеживание айсбергов.

#### 4.1.2. Основные функциональные возможности

Polar View предоставляет мониторинг морского льда в Арктике и Южном океане, просмотр SAR изображений, национальных ледовых карт NIC и met.no, данных AMSR2 по концентрации льда, а также доступ к историческим данным ледовой концентрации. Для пользователя доступны отдельные просмотрщики для Арктики и Антарктики, а также упрощённый режим.

#### 4.1.3. Технические характеристики

С публичной стороны Polar View выглядит как специализированный просмотрщик с веб-интерфейсом, несколькими наборами ледовых слоёв и режимами представления. Официально заявлены стандартный интерфейс и отдельная упрощённая low-bandwidth версия, где набор данных сокращён до ключевых слоёв: SAR high-res, sea ice concentration, ice charts и sea ice edge. SAR в контексте Polar View прямо описывается как радиолокационная съёмка, работающая независимо от облачности и времени суток. Требования со стороны пользователя минимальны: браузер и соединение с интернетом; подробные серверные требования или стек backend в открытых материалах не раскрыты.

#### 4.1.4. Преимущества

- предметная ориентированность (создан для морской безопасности)
- сочетание нескольких источников информации (SAR, ледовые карты, концентрация)
- наличие упрощенного режима (низкая пропускная способность)

#### 4.1.5. Недостатки

- не позиционируется как сервис для сегментации/классификации по ROI

### 4.2. Copernicus Marine / Arctic Sea Ice Analysis and Forecast

#### 4.1.1. Общее описание

Copernicus Marine Service - крупная европейская платформа морских данных и прогнозов. Даёт не только текущие ледовые характеристики, но и прогноз движения и дрейфа.

#### 4.1.2. Основные функциональные возможности

Продукт использует систему Arctic Sea Ice Analysis and Forecast и предоставляет поля концентрации льда, толщины льда, скорости дрейфа, глубины снега, типа льда, возраста льда и альбедо. Система запускается ежедневно, ассимилирует ручные ледовые карты и толщину льда CS2SMOS зимой и выдаёт 9-дневный прогноз. В пользовательском отношении это мощный источник готовых прогностических ледовых полей, а не интерактивный сервис пиксельной сегментации.

#### 4.1.3. Технические характеристики

Технологическое ядро продукта - модель neXtSIM с реологией Brittle-Bingham-Maxwell, работающая на адаптивной треугольной сетке со средним размером ячейки 10 км. Модель принудительно задаётся атмосферными воздействиями ECMWF и океаническими воздействиями TOPAZ5. Продукт поставляет среднемесячные и среднечасовые срезы в виде готовых научно-оперативных данных. Для конечного пользователя локальная вычислительная инфраструктура не обязательна: можно работать через веб-доступ и скачивание продуктов, а тяжёлая модельная часть находится на стороне провайдера.

#### 4.1.4. Преимущества

- Содержит и учитывает скорость движения, тип, возраст, толщину льда
- Ежедневно обновляется
- Может прогнозировать на 9 дней вперед

#### 4.1.5. Недостатки


Платформа сильна как источник данных и научно-операционный сервис, но не как простой интерфейс для интерактивного ROI анализа. Кроме того, пространственная детализация здесь определяется моделью, а не высокодетальными сценами отдельных спутниковых снимков.

### 4.3. Sentinel Hub EO Browser

#### 4.1.1. Общее описание

EO Browser - облачный веб-инструмент для визуализации и анализа спутниковых данных из множества коллекций. Это один из лучших аналогов не по ледовой специализации, а по пользовательскому паттерну работы: выбор области интереса, подбор сцен, кастомная визуализация, измерения, статистика, выгрузка и таймлапсы.

#### 4.1.2. Основные функциональные возможности

EO Browser позволяет мгновенно визуализировать данные из разных спутниковых коллекций, автоматически подбирая сцены, выполняя обработку и создание мозайки. Пользователь может рисовать прямоугольный или полигональный AOI либо загружать свой AOI, измерять площадь и расстояние, создавать timelapse, сравнивать сцены, выполнять экспорт в JPG, PNG, KMZ и GeoTIFF, а также строить статистические графики. Для статистики доступны графики по времени для точки или полигона и гистограммы, если слой поддерживает статистический API через evalscript.

#### 4.1.3. Технические характеристики

Сервис полностью облачный: фоновые операции по выбору подходящих сцен, скачиванию, обработке и мозаицированию выполняются на стороне Sentinel Hub. Поддерживаются готовые визуализации, пользовательские сценарии отображения и слияния данных, включая совместное использование, например, Sentinel-1 и Sentinel-2. Важная деталь: статистические функции зависят от корректно подготовленного evalscript и Statistical API, а timelapse ограничен максимум 300 кадрами за один запуск.

#### 4.1.4. Преимущества

Сильнейшая сторона EO Browser — UX для работы с областью интереса: выделение AOI, сравнение, экспорт, статистика и таймсерии реализованы очень удобно. Второй плюс — мультисенсорность и возможность быстро работать с разными спутниковыми источниками без ручной подготовки сцен. Третий плюс — поддержка high-resolution и даже коммерческих VHR-источников в едином интерфейсе.


#### 4.1.5. Недостатки

EO Browser не является специализированной ледовой системой и не предлагает из коробки предметную классификацию типов льда или прогноз движения льда. Кроме того, часть продвинутых возможностей требует понимания evalscript и особенностей статистического API, то есть порог входа выше, чем у полностью предметного приложения.


### 4.4. ArcGIS Image (Online/Server)

#### 4.1.1. Общее описание

ArcGIS Image - это enterprise/SaaS-платформа Esri для хранения, публикации, анализа и deep learning-обработки растровых и спутниковых данных. Из всех аналогов это, пожалуй, самый сильный универсальный инструмент: здесь уже есть классификация пикселей, детекция изменений, зональная статистика, слои удаленного хранения и интеграция результатов в дашборды и web приложения.

#### 4.1.2. Основные функциональные возможности

Платформа поддерживает классификацию изображений, детекцию изменений, анализ пригодности (для принятия решений), многомерный анализ (включая временной разрез) и продвинутые пользовательские потоки с использованием моделей глубокого обучения. Инструмент Classify Pixels Using Deep Learning позволяет применять обученную модель к слою и создавать удаленный слой на выходе. Полученные классифицированные тематические слои можно использовать для детекции категориальных изменений и их количественного анализа. Также поддерживаются удаленные тайловые слои и динамические растровые слои (обработка на лету), серверная аналитика растров и зональная статистика.

#### 4.1.3. Технические характеристики

ArcGIS Image работает в двух основных режимах: как SaaS в ArcGIS Online и как enterprise-развёртывание через ArcGIS Image Server. В ArcGIS Online анализ требует подходящего типа пользователа - Professional или Professional Plus - и роли с привилегией Imagery Analysis; публикация удаленных слоев требует Creator, Professional или Professional Plus и соответствующих прав. Архитектурно поддерживаются тайловые слои, где отрисовка идёт на клиенте, и динамические растровые слои, где обработка и визуализация идут на сервере. Для растровой аналитики с использованием глубокого обучения в ArcGIS Image Server нужно развёртывание Raster Analysis server, установка библиотек для глубокого обучения и, при использовании GPU, NVIDIA GPU с CUDA compute capability не ниже 5.0, драйвером не ниже 527.41, минимум 6 GB VRAM и рекомендованными 16+ GB. Также поддерживается использование нескольких GPU.

#### 4.1.4. Преимущества

- почти максимальная полнота функций для корпоративной геоаналитики: хранение, публикация, инференс, детекция изменений, зональная статистика, интеграция в карты, приложения и дэшборды
- масштабируется
- возможно использовать собственные модели глубокого обучения и получать удаленный слой как стандартизованный результат

#### 4.1.5. Недостатки

Высокий инфраструктурный и организационный порог входа. Это тяжёлый enterprise-стек с лицензированием, ролями, привилегиями и, в self-hosted сценарии, отдельными требованиями к GPU и серверной конфигурации. Для узкой прикладной задачи “ледовая сегментация и прогноз в веб-интерфейсе” такая система может оказаться избыточной и перегруженной по UX. Кроме того, это универсальная геоплатформа, а не специализированный ледовый продукт.


### 4.5. IceNet

#### 4.1.1. Общее описание

IceNet - ИИ экосистема с открытым исходным кодом для прогнозирования морского льда. Проект разрабатывается международной командой под руководством British Antarctic Survey и The Alan Turing Institute и ориентирован именно на прогноз концентрации льда, а не на готовый картографический интерфейс для конечного пользователя.

#### 4.1.2. Основные функциональные возможности

IceNet позиционируется как экосистема, способная загружать, обрабатывать, тренировать и предсказывать. Пользователь может работать либо через Python, либо через терминал. По официальным материалам система сейчас формирует ежедневные прогнозы концентрации льда до двух недель вперёд и средневзвешенные месячные прогнозы до шести месяцев вперёд. В документации отдельно показан pipeline-подход от настройки до визуализации.

#### 4.1.3. Технические характеристики

IceNet распространяется как Python-пакет и ставится через pip; также возможна установка из исходников. Архитектурно это не готовое web-приложение, а библиотека и CLI-экосистема, причём разработчики отдельно предупреждают, что проект находится в активной разработке и стабильность API не гарантирована.

#### 4.1.4. Преимущества

Сильнейшее преимущество IceNet - открытость и фокус именно на прогнозе морского льда. Для исследовательской или прикладной разработки это очень полезный ориентир: можно использовать подходы IceNet как эталон, как источник архитектурных идей и как основу для собственного предсказательного бекенда. Плюсом также является поддержка полного цикла - от загрузки данных до обучения и предсказания.

#### 4.1.5. Недостатки

- отсутствие полноценного user-friendly web UI для конечного аналитика
- активная стадия разработки и нестабильность API
- установка и эксплуатация требуют Python-окружения, зависимостей TensorFlow/CUDA и общей DevOps-подготовки

#### Разрабатываемая система

**Strengths (Сильные стороны)**

- Узкая предметная специализация. Система ориентирована именно на анализ ледовой обстановки, а не на общий просмотр спутниковых данных.

- Простой и доступный web-интерфейс. Streamlit позволяет быстро предоставить пользователю понятный интерфейс без сложного GIS-стека.

- ROI-first подход. Пользователь работает сразу с интересующей областью, что делает анализ прикладным и удобным.

- Объединение нескольких задач в одном интерфейсе. Выделение области, сегментация, классификация и статистика выполняются в едином workflow.

- Низкий порог входа для пользователя. Не требуется знание сложных GIS-инструментов или ML-фреймворков.

- Быстрая разработка и прототипирование. Streamlit хорошо подходит для быстрого вывода MVP и апробации идеи.

**Weaknesses (Слабые стороны)**

- Ограниченность начального функционала. На первом этапе система решает задачу анализа текущего состояния льда, но не прогнозирует его движение и изменения во времени.

- Зависимость от качества входных данных. Точность сегментации и классификации зависит от типа снимков, их разрешения, шумов, облачности и сезона.

- Ограниченная масштабируемость Streamlit при росте нагрузки. При увеличении числа пользователей или тяжёлых вычислений потребуется выделение backend-сервисов.

- Сложность валидации результатов. Для подтверждения качества нужны размеченные датасеты и экспертные эталоны по типам льда.

- Ограниченная интерпретируемость ML-модели. Пользователю трудно понять причины классификации.

- Зависимость от вычислительных ресурсов. При обработке больших изображений и сложных моделей требуется мощный GPU.

**Opportunities (Возможности)**

- Расширение до прогноза движения льда. Переход от статического анализа к динамическому моделированию и визуализации дрейфа.

- Интеграция разных типов данных ДЗЗ. Добавление SAR, мультиспектральных и оптических снимков.

- Автоматизация аналитики. Подсчёт площадей, долей классов, изменений во времени, генерация отчётов.

- Использование в научных и прикладных задачах. Применение в экологии, логистике, судоходстве.

- Модульное развитие архитектуры. API, хранение результатов, история анализов, многопользовательский режим.

- Интеграция с внешними ледовыми и метеоданными. Повышение точности и ценности системы.

**Threats (Угрозы)**

- Конкуренция со специализированными ледовыми сервисами. Пользователи могут предпочесть уже существующие решения.

- Нестабильность или ограниченная доступность данных. Проблемы с качеством и покрытием спутниковых данных.

- Высокие ожидания пользователей к точности. Ошибки могут снизить доверие.

- Изменчивость ледовых условий. Сезонность и климатические факторы ухудшают переносимость моделей.

- Технические риски внедрения. Проблемы с производительностью при больших объёмах данных.

- Появление более мощных решений. Конкуренция с развивающимися GeoAI-платформами.

## 8. ВЫЯВЛЕННЫЕ ПРОБЛЕМЫ И ОГРАНИЧЕНИЯ СУЩЕСТВУЮЩИХ РЕШЕНИЙ

### 8.1. Фрагментированность функциональности

**Описание проблемы.**
Современные решения покрывают отдельные аспекты задачи, но не обеспечивают полный аналитический цикл. Одни системы ориентированы на предоставление данных (Copernicus Marine, Polar View), другие — на обработку и визуализацию (EO Browser, ArcGIS), третьи — на прогнозирование (IceNet). Однако ни одна из систем не объединяет все этапы: выбор ROI, сегментацию, классификацию, статистику и прогноз.

**Влияние на пользователей.**
Пользователю приходится использовать несколько инструментов:

- получать данные в одной системе,

- обрабатывать их в другой,

- анализировать и визуализировать в третьей.

Это увеличивает:

- время анализа,

- вероятность ошибок,

- требования к квалификации пользователя.

**Распространённость.**
Данная проблема является системной и характерна практически для всех рассмотренных решений. Она обусловлена историческим развитием систем как узкоспециализированных инструментов.

### 8.2. Отсутствие ROI-first подхода

**Описание проблемы.**
Большинство систем построены вокруг данных (data-centric), а не задач пользователя. Пользователь вынужден сначала выбирать источник данных, параметры съёмки и продукты, вместо того чтобы начать с области интереса (ROI).

**Влияние на пользователей.**
Это приводит к:

- увеличению когнитивной нагрузки,

- усложнению интерфейса,

- снижению скорости работы.

Для прикладных пользователей (например, судоходства или мониторинга) это особенно критично, так как их задача — анализ конкретной территории, а не работа с каталогами данных.

**Распространённость.**
Проблема широко распространена:

- выражена в Copernicus Marine,

- частично решена в EO Browser,

практически отсутствует в специализированных ледовых системах как ключевой сценарий.

### 8.2. Отсутствие интегрированной статистики по ROI

**Описание проблемы.**
В большинстве систем статистический анализ либо отсутствует, либо вынесен в отдельные инструменты. Даже при наличии ROI, пользователь не всегда может автоматически получить:

- площади классов,

- доли типов льда,

- агрегированные показатели.

**Влияние на пользователей.**
Пользователь вынужден:

- вручную экспортировать данные,

- обрабатывать их в сторонних инструментах (например, Python, GIS),

- выполнять расчёты отдельно.

Это увеличивает время анализа и снижает удобство.

**Распространённость.**
Проблема:

- ярко выражена в Polar View и Copernicus,

- частично решена в EO Browser,

- решена в ArcGIS, но требует сложной настройки.

### 8.2. Отсутствие специализированных решений “под лед” с полным циклом анализа

**Описание проблемы.**
Существует разрыв между:

- специализированными ледовыми сервисами (данные),

-  универсальными GeoAI-платформами (инструменты).

**Влияние на пользователей.**
Пользователь не получает:

- удобный инструмент именно для ледовой аналитики,

- полный цикл анализа в одном месте,

- специализированные метрики и визуализации.

**Распространённость.**
Это ключевая проблема, характерная для всей предметной области.

## 9. КОНКУРЕНТНЫЕ ПРЕИМУЩЕСТВА РАЗРАБАТЫВАЕМОГО РЕШЕНИЯ

### 9.1. Единый аналитический workflow

Описание:
Система объединяет в одном интерфейсе полный цикл анализа: выделение ROI, сегментацию, классификацию типов льда и расчет статистики.

Обоснование:
Существующие решения фрагментированы: пользователь вынужден работать с несколькими инструментами для выполнения одной задачи.

Ценность для пользователя:
Сокращение времени анализа, снижение количества ошибок и упрощение рабочего процесса.

Отличие от конкурентов:
В аналогах функции распределены между разными системами: данные — в Copernicus, визуализация — в EO Browser, анализ — в ArcGIS, прогноз — в IceNet.

### 9.2. ROI-first подход

Описание:
Анализ начинается с выбора области интереса, а не с выбора данных или продукта.

Обоснование:
Пользователь в прикладных задачах мыслит географически (по области), а не по типу данных.

Ценность для пользователя:
Интуитивно понятный процесс работы и снижение когнитивной нагрузки.

Отличие от конкурентов:
Большинство систем используют data-centric подход; полноценный ROI-first реализован только частично (например, в EO Browser).

### 9.1. Специализация на анализе ледовой обстановки

Описание:
Система ориентирована на сегментацию и классификацию именно типов льда.

Обоснование:
Универсальные платформы не учитывают специфику ледовой типологии и процессов.

Ценность для пользователя:
Более точные и релевантные результаты для задач судоходства, мониторинга и исследований.

Отличие от конкурентов:
ArcGIS и EO Browser универсальны, Polar View — специализирован, но не предоставляет пользовательской аналитики.

### 9.1. Автоматический расчет статистики по ROI

Описание:
После сегментации система автоматически рассчитывает статистику: площади, доли классов и другие показатели.

Обоснование:
В большинстве систем статистика требует отдельной обработки.

Ценность для пользователя:
Мгновенное получение количественных характеристик без дополнительной работы.

Отличие от конкурентов:
В EO Browser статистика ограничена, в Polar View отсутствует, в ArcGIS требует настройки и дополнительных шагов.

### 9.1. Низкий порог входа

Описание:
Использование Streamlit позволяет создать простой и понятный web-интерфейс.

Обоснование:
Многие существующие системы требуют знаний GIS, API или программирования.

Ценность для пользователя:
Система доступна широкому кругу пользователей без специальной подготовки.

Отличие от конкурентов:
ArcGIS и IceNet требуют профессиональных навыков, EO Browser — частично.

### 9.1. Возможность интеграции различных источников данных

Описание:
Система может работать с различными типами спутниковых данных (SAR, оптические и др.).

Обоснование:
Разные типы данных дают различную информацию о ледовой обстановке.

Ценность для пользователя:
Повышение точности анализа и устойчивости к условиям съёмки (например, облачности).

Отличие от конкурентов:
В специализированных системах набор данных фиксирован, в универсальных — интеграция сложнее.

- автоматизировать анализ ледовой обстановки

- высокая вовлечённость

- готовность к работе с новыми инструментами

- понимание предметной области

### 10.2. Вторичная целевая аудитория

Вторичная аудитория — пользователи, которым важны результаты анализа, но которые не являются специалистами в области ДЗЗ или машинного обучения.

**Состав аудитории**

- специалисты морской логистики и судоходства

- операторы арктических и антарктических маршрутов

- сотрудники государственных структур

- специалисты по управлению рисками

- экологические организации

- образовательные учреждения (студенты, преподаватели)

**Потребности**

- простая и понятная визуализация ледовой обстановки

- быстрое получение информации по конкретной территории

- поддержка принятия решений

- минимальные требования к технической подготовке

**Ценность сервиса для аудитории**

Сервис обеспечивает:

- быстрый доступ к аналитике без необходимости глубоких знаний

- наглядное представление информации

- снижение рисков при принятии решений

- возможность использовать систему как инструмент поддержки

**Особенности использования**

- нерегулярное или ситуативное использование

- низкий уровень технической подготовки

- высокая потребность в простоте интерфейса

- ориентация на результат, а не на процесс анализа

## 11. ВЫВОДЫ

### 11.1. Общие выводы по анализу рынка

Специализированные ледовые сервисы (например, Polar View), ориентированные на предоставление готовых данных и визуализацию ледовой обстановки.

Универсальные геоинформационные и EO-платформы (Sentinel Hub EO Browser, ArcGIS Image), обеспечивающие инструменты анализа и обработки данных.

Научные и AI-ориентированные решения (IceNet, Copernicus Marine), направленные на моделирование и прогнозирование ледовых процессов.

Несмотря на высокий уровень развития каждого из классов, анализ показал, что существующие решения обладают фрагментированной функциональностью. Они, как правило, решают отдельные задачи, но не обеспечивают полный цикл анализа ледовой обстановки в рамках единого пользовательского интерфейса.

**Ключевыми недостатками существующих решений являются:**

- отсутствие интеграции всех этапов анализа (ROI → сегментация → классификация → статистика → прогноз),

- сложность использования для прикладных пользователей,

- высокая зависимость от внешних инструментов,

- недостаточная адаптация под конкретные предметные задачи (ледовая обстановка).

Таким образом, на рынке отсутствует компактное и удобное решение, ориентированное на комплексный анализ ледовой обстановки в рамках одного сервиса.

### 11.2. Обоснование актуальности проекта

С технологической точки зрения, развитие методов машинного обучения и доступность спутниковых данных создают предпосылки для автоматизации анализа ледовой обстановки. Однако существующие системы либо не используют эти возможности в полной мере, либо делают их недоступными для широкого круга пользователей.

С прикладной точки зрения, задачи анализа ледового покрова имеют важное значение для:

- судоходства и морской логистики,

- экологического мониторинга,

- научных исследований,

- оценки климатических изменений.

При этом пользователи сталкиваются с рядом проблем:

- сложность работы с существующими системами,

- необходимость использования нескольких инструментов,

- отсутствие удобных средств анализа конкретных областей.

Разрабатываемое решение направлено на устранение этих проблем за счёт:

- интеграции ключевых функций в одном интерфейсе,

- упрощения пользовательского взаимодействия,

- автоматизации аналитических процессов.

Таким образом, проект является актуальным и отвечает современным требованиям к инструментам обработки геопространственных данных.

### 11.3. Позиционирование разрабатываемого решения

Разрабатываемая система занимает промежуточное положение между специализированными ледовыми сервисами и универсальными геоинформационными платформами.


**С одной стороны, она наследует от специализированных решений:**

- ориентацию на предметную область (ледовая обстановка),

- использование тематических данных и моделей.  

**С другой стороны, она заимствует у универсальных платформ:**

- инструменты обработки и анализа данных,

- интерактивный веб-интерфейс,

- возможность работы с ROI.

**При этом ключевое отличие заключается в том, что система реализует полный аналитический цикл в рамках одного сервиса, включая:**

- выделение области интереса,

- сегментацию ледового покрова,

- классификацию типов льда,

- расчет статистических показателей.

**Таким образом, система позиционируется как:**

специализированный веб-сервис для комплексного анализа ледовой обстановки с использованием методов машинного обучения

**В конкурентной среде решение занимает нишу:**

- более простого и доступного инструмента по сравнению с enterprise-системами,

- более функционального и аналитически насыщенного по сравнению с классическими viewer-сервисами.

### 11.4. Риски и ограничения

**Технические риски (точность моделей)**

**Описание:**
Качество сегментации и классификации льда напрямую зависит от обученных моделей машинного обучения.

**Проблемы:**

- ошибки классификации типов льда,

- ухудшение качества при смене региона или сезона,

- чувствительность к шумам (SAR speckle, облачность).

**Влияние:**
Снижение доверия пользователей и ограничение практического применения.

**Возможные решения:**

- использование разнообразных обучающих датасетов (разные регионы и сезоны),

- применение data augmentation,

- ансамбли моделей (например, CNN + Transformer),

- регулярная валидация на эталонных данных,

- добавление оценки уверенности модели.










**Ограничения данных**

**Описание:**
Спутниковые данные имеют ограничения по качеству, разрешению и доступности.

**Проблемы:**

- облачность (для оптических данных),

- шумы в SAR-изображениях,

- различия в разрешении и частоте съёмки.

**Влияние:**
- Снижение точности анализа и нестабильность результатов.

**Возможные решения:**

- комбинирование источников данных (SAR + оптические),

- предварительная обработка (фильтрация шума, нормализация),

- выбор оптимальных источников под задачу,

- кэширование и накопление исторических данных.

**Конкурентные риски**

**Описание:**
Существующие платформы могут расширить функциональность.

**Проблемы:**

- появление аналогичных функций в EO или GIS-системах,

- усиление конкуренции со стороны крупных игроков.

**Влияние:**
- Снижение уникальности и конкурентных преимуществ.

**Возможные решения:**

- фокус на узкой специализации (ледовая аналитика),

- развитие уникальных функций (ROI-first + статистика),

- ускорение разработки и внедрения новых возможностей,

- ориентация на конкретные пользовательские сценарии.
